Solutions for Tasks on Maryland Traffic Violations Data Set

To start, let's load the data.  In this case, there are quite a few columns and not all of them will be of interest for the task solutions.

```python
Traffic_Violation = pd.read_csv(r'/kaggle/input/traffic-violations-in-maryland-county/Traffic_Violations.csv')
Traffic_Violation.head()
```

```python
Traffic_Violation.columns
```

```python
Traffic_Violation2 = Traffic_Violation.drop(columns=['Date Of Stop', 'Time Of Stop', 'Agency', 'SubAgency', 'Description',
       'Location', 'Latitude', 'Longitude', 'Accident', 
        'Fatal',
       'HAZMAT', 'State',
       'VehicleType', 'Year', 'Make', 'Model', 'Color',
       'Charge', 'Article', 'Contributed To Accident', 'Race', 'Gender',
       'Driver City', 'Driver State', 'DL State', 'Arrest Type',
       'Geolocation'])
```

```python
Traffic_Violation2 = Traffic_Violation2.dropna()
```

```python
Traffic_Violation2.info()
```

```python


Traffic_Violation2['Belts']=Traffic_Violation2['Belts'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Belts']=Traffic_Violation2['Belts'].replace(to_replace='No',value='0')

Traffic_Violation2['Personal Injury']=Traffic_Violation2['Personal Injury'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Personal Injury']=Traffic_Violation2['Personal Injury'].replace(to_replace='No',value='0')

Traffic_Violation2['Property Damage']=Traffic_Violation2['Property Damage'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Property Damage']=Traffic_Violation2['Property Damage'].replace(to_replace='No',value='0')

Traffic_Violation2['Commercial License']=Traffic_Violation2['Commercial License'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Commercial License']=Traffic_Violation2['Commercial License'].replace(to_replace='No',value='0')

Traffic_Violation2['Commercial Vehicle']=Traffic_Violation2['Commercial Vehicle'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Commercial Vehicle']=Traffic_Violation2['Commercial Vehicle'].replace(to_replace='No',value='0')

Traffic_Violation2['Alcohol']=Traffic_Violation2['Alcohol'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Alcohol']=Traffic_Violation2['Alcohol'].replace(to_replace='No',value='0')

Traffic_Violation2['Work Zone']=Traffic_Violation2['Work Zone'].replace(to_replace='Yes',value='1')
Traffic_Violation2['Work Zone']=Traffic_Violation2['Work Zone'].replace(to_replace='No',value='0')

Traffic_Violation2['Violation Type']=Traffic_Violation2['Violation Type'].replace(to_replace='Citation',value='1')
Traffic_Violation2['Violation Type']=Traffic_Violation2['Violation Type'].replace(to_replace='Warning',value='0')
Traffic_Violation2['Violation Type']=Traffic_Violation2['Violation Type'].replace(to_replace='ESERO',value='2')
```

```python
Traffic_Violation2.head(5)
```

```python
Traffic_Violation2['Belts'] = Traffic_Violation2['Belts'].astype(int)
Traffic_Violation2['Personal Injury'] = Traffic_Violation2['Personal Injury'].astype(int)
Traffic_Violation2['Property Damage'] = Traffic_Violation2['Property Damage'].astype(int)

Traffic_Violation2['Commercial License'] = Traffic_Violation2['Commercial License'].astype(int)

Traffic_Violation2['Commercial Vehicle'] = Traffic_Violation2['Commercial Vehicle'].astype(int)
Traffic_Violation2['Alcohol'] = Traffic_Violation2['Alcohol'].astype(int)
Traffic_Violation2['Work Zone'] = Traffic_Violation2['Work Zone'].astype(int)
```

```python
Traffic_Violation2.head(5)
```

```python
X = Traffic_Violation2.drop(columns=['Violation Type'])
y = Traffic_Violation2['Violation Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
classifier = tree.DecisionTreeClassifier(max_depth=5)
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train, y_train)
```

```python
import graphviz 
tree_data = tree.export_graphviz(classifier, out_file=None,
                               feature_names=['Belts','Personal Injury','Property Damage',
                                              'Commercial License','Commercial Vehicle',
                                              'Work Zone','Alcohol'],
                               class_names=['Yes','No'],filled=True,
                                rounded=True,  
                              special_characters=True) 
graph = graphviz.Source(tree_data)
graph.render("Gini")
graph

```
